```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset again
rsu_data_path = "RSU_Interactions.csv"
rsu_data = pd.read_csv(rsu_data_path)

# Calculate the latency for each interaction
rsu_data['Latency'] = rsu_data['Receive Time'] - rsu_data['Send Time']

# Plot the histogram of latencies
plt.figure(figsize=(10, 6))
plt.hist(rsu_data['Latency'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Latency (s)')
plt.ylabel('Frequency')
plt.title('Distribution of Latencies Across RSU Interactions')
plt.grid(True, linestyle='--')
plt.show()
```

```python
# Creating a pivot table to represent the interaction frequencies between different RSUs
interaction_matrix = rsu_data.pivot_table(index='Sender RSU ID', columns='Receiver RSU ID', aggfunc=len, fill_value=0, values='Send Time')

# Displaying the first few rows of the interaction matrix
interaction_matrix.head()
```

```python
import seaborn as sns

# Plotting the heat map using Seaborn
plt.figure(figsize=(12, 10))
sns.heatmap(interaction_matrix, cmap='viridis', annot=False)
plt.title('Heat Map of RSU Interaction Frequencies')
plt.xlabel('Receiver RSU ID')
plt.ylabel('Sender RSU ID')
plt.show()
```
