```python
# Importing necessary libraries
import pandas as pd

# Reading the CSV file
file_path = "../master_final.csv"
data = pd.read_csv(file_path)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np

# Extracting the hour of the day by splitting the arrival_time_new string
data['hour_of_day'] = data['arrival_time_new'].apply(lambda x: int(x.split()[2].split(':')[0])) 

# Define the features and target variables
selected_features = ['stop_sequence', 'day_of_the_week', 'time_category', 'dist_to_next_stop', 'hour_of_day']
selected_targets = ['next_stop_sequence', 'time_diff']

# Prepare the input and output arrays
X = data[selected_features]
y = data[selected_targets]

# One-hot encode the categorical variables
categorical_features = ['day_of_the_week', 'time_category']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", 
                                  one_hot, 
                                  categorical_features)], 
                                remainder="passthrough")
X_transformed = transformer.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# Creating the Random Forest regressor
rf_model = MultiOutputRegressor(RandomForestRegressor(random_state=42))

# Training the Random Forest model on the training data
rf_model.fit(X_train, y_train)

# Predicting on the testing set
rf_y_pred = rf_model.predict(X_test)

# Evaluating the Random Forest model using MAE, RMSE, and RÂ² score
rf_mae_next_stop_sequence = mean_absolute_error(y_test['next_stop_sequence'], rf_y_pred[:, 0])
rf_rmse_next_stop_sequence = np.sqrt(mean_squared_error(y_test['next_stop_sequence'], rf_y_pred[:, 0]))
rf_r2_next_stop_sequence = r2_score(y_test['next_stop_sequence'], rf_y_pred[:, 0])

rf_mae_time_diff = mean_absolute_error(y_test['time_diff'], rf_y_pred[:, 1])
rf_rmse_time_diff = np.sqrt(mean_squared_error(y_test['time_diff'], rf_y_pred[:, 1]))
rf_r2_time_diff = r2_score(y_test['time_diff'], rf_y_pred[:, 1])

(rf_mae_next_stop_sequence, rf_rmse_next_stop_sequence, rf_r2_next_stop_sequence), (rf_mae_time_diff, rf_rmse_time_diff, rf_r2_time_diff)
```

```python
# 1. Filtering the rows from the original dataset where dist_to_next_stop is between 200 and 280 meters
filtered_data_X = data[(data['dist_to_next_stop'] >= 200) & (data['dist_to_next_stop'] <= 280)][selected_features]

# 2. Extracting the corresponding y values
filtered_data_y = data.loc[filtered_data_X.index, selected_targets]

# Re-fitting the transformer on the selected_features of the entire dataset
transformer.fit(data[selected_features])

# Transforming the filtered_data_X again using the re-fitted transformer
filtered_X_transformed_corrected = transformer.transform(filtered_data_X)

# Predicting the time_diff for the transformed filtered data
filtered_y_pred_rf_corrected = rf_model.predict(filtered_X_transformed_corrected)[:, 1]

# Calculating the residuals
residuals_rf_corrected = filtered_y_pred_rf_corrected - filtered_data_y['time_diff'].values

# Summary statistics for the residuals
residual_summary_rf_corrected = pd.Series(residuals_rf_corrected).describe()
residual_summary_rf_corrected
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the plot
plt.figure(figsize=(14, 6))

# Histogram for the residuals
plt.subplot(1, 2, 1)
sns.histplot(residuals_rf_corrected, bins=20, kde=True)
plt.title('Histogram of Residuals')
plt.xlabel('Residuals (time_diff)')
plt.ylabel('Frequency')

# Box plot for the residuals
plt.subplot(1, 2, 2)
sns.boxplot(x=residuals_rf_corrected)
plt.title('Box Plot of Residuals')
plt.xlabel('Residuals (time_diff)')

plt.tight_layout()
plt.show()
```
